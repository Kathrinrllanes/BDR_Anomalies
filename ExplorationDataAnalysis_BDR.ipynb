```python
#Change from string to date type 
from datetime import datetime

def to_date (date_str):
    return datetime.strptime(date_str, '%d%b%Y').date()

#pd_BDR_mod['Fecha_Operac'] = pd_BDR['Fecha_Operac'].apply(lambda x: to_date(x), axis=1)
pd_BDR['Fecha_Operac'] = pd_BDR['Fecha_Operac'].apply(lambda x: to_date(x))
pd_BDR['Fecha_Inicio'] = pd_BDR['Fecha_Inicio'].apply(lambda x: to_date(x))
pd_BDR['Fecha_Baja'] = pd_BDR['Fecha_Baja'].apply(lambda x: to_date(x))
pd_BDR['Fecha_VCTO'] = pd_BDR['Fecha_VCTO'].apply(lambda x: to_date(x))
pd_BDR
```

```python
from datetime import datetime
from pyspark.sql.functions import col, udf
from pyspark.sql.types import DateType

# This function converts the string cell into a date:
func =  udf (lambda x: datetime.strptime(x, '%d%b%Y'), DateType())

#Apply the function to columns
df_BDR = df_BDR.withColumn("Fecha_Operac", func(col("Fecha_Operac")))
df_BDR = df_BDR.withColumn("Fecha_Inicio", func(col("Fecha_Inicio")))
df_BDR = df_BDR.withColumn("Fecha_Baja", func(col("Fecha_Baja")))
df_BDR = df_BDR.withColumn("Fecha_VCTO", func(col("Fecha_VCTO")))


df_BDR.show(3)

df_BDR.printSchema()
```
